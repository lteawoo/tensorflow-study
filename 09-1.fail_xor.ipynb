```python
import numpy as np
import tensorflow as tf

x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y_data = np.array([[0], [1], [1], [0]], dtype=np.float32)

X = tf.constant(x_data)
Y = tf.constant(y_data)
W = tf.Variable(tf.random.normal([2, 1], name='weight'))
b = tf.Variable(tf.random.normal([1]), name='bias')

def hypothesis():
    return tf.math.sigmoid(tf.linalg.matmul(X, W) + b)

def cost():
    return -tf.reduce_mean(Y * tf.math.log(hypothesis()) + (1 - Y) * tf.math.log(1 - hypothesis()))

def train():
    return tf.keras.optimizers.SGD(learning_rate = 1e-5).minimize(cost, var_list=[W, b])

def predicted():
    return tf.cast(hypothesis() > 0.5, dtype=tf.float32)

def accuracy():
    return tf.reduce_mean(tf.cast(tf.equal(predicted(), Y), dtype=tf.float32))

for step in range(10001):
    train()
    if step % 100 == 0:
        print(step, cost())
```

```python
h = hypothesis()
c = predicted()
a = accuracy()
print('hypothesis: ', h.numpy(), 'Correct: ', c.numpy(), 'Accuracy: ', a.numpy())
```
