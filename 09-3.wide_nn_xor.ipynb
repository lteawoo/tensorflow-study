```python
import numpy as np
import tensorflow as tf
```

```python
x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y_data = np.array([[0], [1], [1], [0]], dtype=np.float32)

X = tf.constant(x_data)
Y = tf.constant(y_data)

W1 = tf.Variable(tf.random.normal([2, 10], name='weight1'))
b1 = tf.Variable(tf.random.normal([10]), name='bias1')
def layer1():
    return tf.math.sigmoid(tf.linalg.matmul(X, W1) + b1)

W2 = tf.Variable(tf.random.normal([10, 1], name='weight2'))
b2 = tf.Variable(tf.random.normal([1]), name='bias2')
def hypothesis():
    return tf.math.sigmoid(tf.linalg.matmul(layer1(), W2) + b2)

def cost():
    return -tf.reduce_mean(Y * tf.math.log(hypothesis()) + (1 - Y) * tf.math.log(1 - hypothesis()))

def train():
    return tf.keras.optimizers.SGD(learning_rate = 0.1).minimize(cost, var_list=[W1, b1, W2, b2])

def predicted():
    return tf.cast(hypothesis() > 0.5, dtype=tf.float32)

def accuracy():
    return tf.reduce_mean(tf.cast(tf.equal(predicted(), Y), dtype=tf.float32))

for step in range(10001):
    train()
    if step % 100 == 0:
        print(step, cost())
```

```python
h = hypothesis()
c = predicted()
a = accuracy()
print('hypothesis: ', h.numpy(), 'Correct: ', c.numpy(), 'Accuracy: ', a.numpy())
```
