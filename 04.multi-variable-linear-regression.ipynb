```python
import tensorflow as tf

x1_data = tf.constant([73., 93., 89., 96., 73.])
x2_data = tf.constant([80., 88., 91., 98., 66.])
x3_data = tf.constant([75., 93., 90., 100., 70.])
y_data = tf.constant([152., 185., 180., 196., 142.])

x1 = tf.Variable(1., shape=tf.TensorShape(None), dtype=tf.float32)
x2 = tf.Variable(1., shape=tf.TensorShape(None), dtype=tf.float32)
x3 = tf.Variable(1., shape=tf.TensorShape(None), dtype=tf.float32)
y = tf.Variable(1., shape=tf.TensorShape(None), dtype=tf.float32)

w1 = tf.Variable(tf.random.normal([1]), dtype=tf.float32, name='weight1')
w2 = tf.Variable(tf.random.normal([1]), dtype=tf.float32, name='weight2')
w3 = tf.Variable(tf.random.normal([1]), dtype=tf.float32, name='weight3')
b = tf.Variable(tf.random.normal([1]), dtype=tf.float32, name='bias')

def hypothesis():
    return x1 * w1 + x2 * w2 + x3 * w3 + b

def cost():
    return tf.reduce_mean(tf.square(hypothesis() - y))

def optimizer():
    return tf.keras.optimizers.SGD(learning_rate = 1e-5)

def train():
    return optimizer().minimize(cost, var_list=[w1, w2, w3])

for step in range(2001):
    x1.assign(x1_data)
    x2.assign(x2_data)
    x3.assign(x3_data)
    y.assign(y_data)
    train()
    if (step % 10 == 0):
        print(step, "Cost: ", cost().numpy(), "\nPrediction:\n", hypothesis().numpy())
```

```python
import tensorflow as tf

x_data = tf.constant([
    [73., 80., 75.], [93., 88., 93.], [89., 91., 90.],
    [96., 98., 100.], [73., 66., 70.]
])

y_date = tf.constant([[152.], [185.], [180.], [196.], [142.]])
X = tf.Variable([[1., 1., 1.]], shape=tf.TensorShape(None), dtype=tf.float32)
Y = tf.Variable([[1.]], shape=tf.TensorShape(None), dtype=tf.float32)

W = tf.Variable(tf.random.normal([3, 1]), dtype=tf.float32, name='weight')
b = tf.Variable(tf.random.normal([1]), dtype=tf.float32, name='bias')

def hypothesis():
    return tf.linalg.matmul(X, W) + b

def cost():
    return tf.reduce_mean(tf.square(hypothesis() - y))

def optimizer():
    return tf.keras.optimizers.SGD(learning_rate = 1e-5)

def train():
    return optimizer().minimize(cost, var_list=[W])

for step in range(2001):
    X.assign(x_data)
    Y.assign(y_data)
    train()
    if (step % 10 == 0):
        print(step, "Cost: ", cost().numpy(), "\nPrediction:\n", hypothesis().numpy())
```
